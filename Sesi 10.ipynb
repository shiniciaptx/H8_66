```python
from scipy.stats import uniform
n=10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc=start, scale=width)
```

```python
ax = sns.distplot(data_uniform,
                 bins=100,
                 kde=True,
                 color='skyblue',
                 hist_kws={'linewidth':15,'alpha':1})
ax.set(xlabel='uniform distribution',ylabel='freq')
```

```python
from scipy.stats import norm
data_normal = norm.rvs(size=10000,loc=0,scale=1)

ax = sns.distplot(data_normal,
                 bins=100,
                 kde=True,
                 color='skyblue',
                 hist_kws={'linewidth':15,'alpha':1})
ax.set(xlabel='normal distribution',ylabel='freq')
```

# Diff in Population Means

```python
# populasi : adults in the NHANES data
# parameter of interest : miu1 - miu2, BMI
# H0 : miu1=miu2
# H1 : miu1 != miu2

url="https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/nhanes_2015_2016.csv"
da = pd.read_csv(url)
da.head()
```

```python
females = da[da["RIAGENDR"]==2]
male = da[da["RIAGENDR"]==1]
```

```python
n1 = len(females)
mu1 = females["BMXBMI"].mean()
sd1 = females["BMXBMI"].std()

(n1, mu1, sd1)
```

```python
n2 = len(male)
mu2 = male["BMXBMI"].mean()
sd2 = male["BMXBMI"].std()

(n2, mu2, sd2)
```

```python
sm.stats.ztest(females["BMXBMI"].dropna(),
              male["BMXBMI"].dropna(), alternative='two-sided')
```

```python
plt.figure(figsize=(7,4))
plt.title("Female BMI histogram", fontsize=16)
plt.hist(females["BMXBMI"].dropna(),edgecolor='k',color='pink',bins=25)
plt.show
```

```python
plt.figure(figsize=(7,4))
plt.title("Male BMI histogram", fontsize=16)
plt.hist(male["BMXBMI"].dropna(),edgecolor='k',color='blue',bins=25)
plt.show
```
